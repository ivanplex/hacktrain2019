# NLP

```python
df_1.columns
```

```python
df_2 = df_1[['Section Name', 'Asset Desc 1', 'Asset Desc 2', 'Incident Equipment',
       'Incident Description', 'Incident Reason Name',
       'Incident Category Description', 'ELR Name',
       'Failure Detail-Cause-Action', 'Failure Notes']].copy()
```

```python
df_2.head(10)
```

```python
from fuzzywuzzy import fuzz
```

```python
fuzz.token_sort_ratio('Deluxe Room, 1 King Bed', 'Deluxe King Room')
```

```python
fuzz.token_set_ratio('Deluxe Room, 1 King Bed', 'Deluxe King Room')
```

```python
def get_ratio(row):
    name = row['Asset Desc 2']
    name1 = row['Section Name'] + row['Failure Detail-Cause-Action']
    return fuzz.token_set_ratio(name, name1)

len(df_2[df_2.apply(get_ratio, axis=1) > 70]) / len(df_2)
```

```python
df_2.shape
```

```python
len(df_2['Asset Desc 2'].unique())
```

```python
df_2['Failure Detail-Cause-Action'] = df_2['Failure Detail-Cause-Action'].apply(lambda x: x.replace('[Detail]:', ''))
```

```python
df_2['Text'] = (df_2['Section Name']+' ' + df_2['Failure Detail-Cause-Action']+' ' + df_2['Incident Description']+' ' + 
                  df_2['Incident Reason Name']+' ' + df_2['Failure Notes'])
```

```python
df_2.head(2)
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize as tokenize
import string
stop = stopwords.words('english') + list(string.punctuation)

df_2.Text = df_2.Text.apply(lambda x: ''.join(c for c in x if not c.isnumeric()))
df_2.Text = df_2.Text.apply(lambda x: tokenize(x))
df_2.Text = df_2.Text.apply(lambda x: ' '.join([word for word in x if word not in stop]))
```

```python
df_2.drop_duplicates()
```

```python
X = df_2.Text
y = df_2['Asset Desc 2']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB

from sklearn.pipeline import Pipeline

from sklearn.feature_extraction.text import TfidfTransformer



nb = Pipeline([('vect', CountVectorizer()),

               ('tfidf', TfidfTransformer()),

               ('clf', MultinomialNB()),

              ])

nb.fit(X_train, y_train)


from sklearn.metrics import classification_report

y_pred = nb.predict(X_test)

print('accuracy %s' % accuracy_score(y_pred, y_test))

print(classification_report(y_test, y_pred))
```
